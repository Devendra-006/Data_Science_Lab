```python
#without using any library
# Take user input (numbers separated by space)
data = list(map(float, input("Enter numbers separated by space: ").split()))

n = len(data)

# Mean
mean = sum(data) / n

# Median
data.sort()
if n % 2 == 0:
    median = (data[n//2 - 1] + data[n//2]) / 2
else:
    median = data[n//2]

# Mode
mode = data[0]
max_count = 0
for x in data:
    count = 0
    for y in data:
        if x == y:
            count += 1
    if count > max_count:
        max_count = count
        mode = x

# Variance
variance = 0
for x in data:
    variance += (x - mean) ** 2
variance = variance / n

# Standard Deviation
std_dev = variance ** 0.5

# Display Results
print("\nResults:")
print("Mean =", mean)
print("Median =", median)
print("Mode =", mode)
print("Variance =", variance)
print("Standard Deviation =", std_dev)
```

```python
#Generate a normal distribution and plot its histogram.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Take user input
mean = float(input("Enter mean: "))
std_dev = float(input("Enter standard deviation: "))
size = int(input("Enter number of data points: "))
bins = int(input("Enter number of histogram bins: "))

# Generate normal distribution data
data = np.random.normal(loc=mean, scale=std_dev, size=size)

# Plot histogram
plt.hist(data, bins=bins, density=True)

# Generate x values for curve
x = np.linspace(min(data), max(data), 100)

# Plot normal distribution curve
plt.plot(x, norm.pdf(x, mean, std_dev))

# Labels and title
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.title("Histogram of Normal Distribution")

plt.show()
```

```python
#Simulate a coin toss experiment and compute probabilities.


import numpy as np

# Take user input
tosses = int(input("Enter number of coin tosses: "))

# Simulate coin toss (H = 1, T = 0)
results = np.random.choice([0, 1], size=tosses)

# Count outcomes
heads = np.sum(results == 1)
tails = np.sum(results == 0)

# Compute probabilities
prob_heads = heads / tosses
prob_tails = tails / tosses

# Display results
print("\nResults:")
print("Total Tosses =", tosses)
print("Heads =", heads)
print("Tails =", tails)
print("Probability of Heads =", prob_heads)
print("Probability of Tails =", prob_tails)
```
