```python
#Code to compute the factorial of the number.

def factorial(n):
    if n == 0:
        return 1
    else:
        res = 1
        for i in range(1, n + 1):
            res *= i
        return res

# Get number from user
num = int(input("Enter a non-negative integer: "))

# Calculate and print factorial
if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    print(f"The factorial of {num} is {factorial(num)}")
```

```python
#Code to implement the multiplication table of the number.

# Get number from user
num = int(input("Enter a number: "))

# Display multiplication table
print(f"Multiplication Table for {num}:")
for i in range(1, 11):  # Table from 1 to 10
    print(f"{num} x {i} = {num * i}")
```

```python
#Code to determine that the entered number is odd or even.

# Get number from user
num = int(input("Enter an integer: "))

# Check if the number is odd or even
if num % 2 == 0:
    print(f"The number {num} is Even.")
else:
    print(f"The number {num} is Odd.")
```

```python
#Code to check that the number is palindrome or not

# Get number from user
num_str = input("Enter an integer: ")

# Check if the number is a palindrome
if num_str == num_str[::-1]:
    print(f"The number {num_str} is a Palindrome.")
else:
    print(f"The number {num_str} is not a Palindrome.")
```

```python
#Code to develop the expert system that provides the simpple decision making


print("Welcome to the Simple Expert System!")
print("Let's help you decide what to do today based on the weather.")

# Get user input for weather condition
weather = input("Is the weather 'sunny', 'rainy', or 'cloudy'? ").lower()

# Implement decision-making rules
if weather == 'sunny':
    print("It's sunny! You should go for a walk or a picnic.")
elif weather == 'rainy':
    print("It's rainy. Stay indoors and read a book or watch a movie.")
elif weather == 'cloudy':
    print("It's cloudy. A good day for indoor activities or a light stroll if it doesn't rain.")
else:
    print("I'm not sure what to suggest for that weather condition. Please choose from 'sunny', 'rainy', or 'cloudy'.")
```
