```python
#Code to compute the factorial of the number.

def factorial(n):
    if n == 0:
        return 1
    else:
        res = 1
        for i in range(1, n + 1):
            res *= i
        return res

# Get number from user
num = int(input("Enter a non-negative integer: "))

# Calculate and print factorial
if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    print(f"The factorial of {num} is {factorial(num)}")
```

```python
#Code to implement the multiplication table of the number.

# Get number from user
num = int(input("Enter a number: "))

# Display multiplication table
print(f"Multiplication Table for {num}:")
for i in range(1, 11):  # Table from 1 to 10
    print(f"{num} x {i} = {num * i}")
```

```python
#Code to determine that the entered number is odd or even.

# Get number from user
num = int(input("Enter an integer: "))

# Check if the number is odd or even
if num % 2 == 0:
    print(f"The number {num} is Even.")
else:
    print(f"The number {num} is Odd.")
```

```python
#Code to check that the number is palindrome or not

# Get number from user
num_str = input("Enter an integer: ")

# Check if the number is a palindrome
if num_str == num_str[::-1]:
    print(f"The number {num_str} is a Palindrome.")
else:
    print(f"The number {num_str} is not a Palindrome.")
```

# Task
Create an expert system for an education system that defines relevant facts, formulates decision rules, implements the core logic in Python, allows for user interaction and testing with education scenarios, and summarizes its functionality and potential applications.

## Define Facts for Education System

### Subtask:
Identify and define key pieces of information (facts) relevant to an education system, such as student performance indicators, learning styles, subject interests, and attendance records. These will form the knowledge base.


**Reasoning**:
The subtask requires defining key facts for an education system and storing them in a Python dictionary. The prompt already provides an example of such a dictionary, so I will implement it directly.



```python
knowledge_base = {
    'student_performance': 'good',
    'learning_style': 'visual',
    'subject_interest': 'science',
    'attendance_record': 'regular'
}

print("Knowledge base defined:")
print(knowledge_base)
```

## Formulate Decision Rules for Education

### Subtask:
Create a set of 'if-then' rules based on the defined educational facts. These rules will specify recommendations, interventions, or pathways based on different combinations of student data.


**Reasoning**:
To define the decision rules as specified, I will create a Python list of dictionaries named `rules`, where each dictionary contains 'conditions' and 'recommendation' keys, outlining at least three distinct educational scenarios.



```python
rules = [
    {
        'conditions': {'student_performance': 'low', 'attendance_record': 'poor'},
        'recommendation': 'Recommend academic counseling and parental meeting.'
    },
    {
        'conditions': {'learning_style': 'visual', 'subject_interest': 'science'},
        'recommendation': 'Suggest interactive science workshops and educational videos.'
    },
    {
        'conditions': {'student_performance': 'good', 'attendance_record': 'regular'},
        'recommendation': 'Suggest advanced placement courses or enrichment programs.'
    },
    {
        'conditions': {'student_performance': 'average', 'subject_interest': 'arts'},
        'recommendation': 'Recommend exploring art clubs or creative writing classes.'
    }
]

print("Decision rules defined:")
for i, rule in enumerate(rules):
    print(f"Rule {i+1}: {rule}")
```

## Implement Education Expert System Logic

### Subtask:
Write Python code to implement the core logic of the expert system, including an inference engine that can process the defined rules to provide decisions or recommendations based on the educational facts.


**Reasoning**:
The subtask requires implementing the core logic of the expert system by defining a function that processes rules against a knowledge base and returns recommendations. This will be done in a Python code block.



```python
def run_expert_system(knowledge_base, rules):
    recommendations = []
    for rule in rules:
        conditions_met = True
        for fact, value in rule['conditions'].items():
            if fact not in knowledge_base or knowledge_base[fact] != value:
                conditions_met = False
                break
        if conditions_met:
            recommendations.append(rule['recommendation'])
    return recommendations

# Run the expert system
applicable_recommendations = run_expert_system(knowledge_base, rules)

# Print the recommendations
print("Applicable Recommendations based on the knowledge base:")
if applicable_recommendations:
    for rec in applicable_recommendations:
        print(f"- {rec}")
else:
    print("No specific recommendations based on the current knowledge base.")
```

## User Interaction and Testing for Education Scenarios

### Subtask:
Develop a mechanism for users (e.g., educators, students, parents) to input relevant educational facts. Test the system with various student profiles and scenarios to evaluate its decision-making accuracy and refine the rules as necessary.


### User Interaction and Testing for Education Scenarios

To allow users to interact with the expert system, we will prompt them to input values for various student attributes. This will create a dynamic knowledge base based on user input, which will then be fed into our `run_expert_system` function to generate recommendations. This process will enable testing of the system with diverse educational scenarios.

**Reasoning**:
The next step is to implement the user interaction mechanism as outlined in the subtask instructions. This involves prompting the user for input, constructing a `user_knowledge_base`, running the expert system, and displaying the results.



```python
print("\n--- Enter Student Profile Facts ---")
user_knowledge_base = {}

# Define the possible values for each fact to guide the user
fact_definitions = {
    'student_performance': ['good', 'average', 'low'],
    'learning_style': ['visual', 'auditory', 'kinesthetic'],
    'subject_interest': ['science', 'math', 'arts', 'history'],
    'attendance_record': ['regular', 'poor']
}

for fact, possible_values in fact_definitions.items():
    while True:
        user_input = input(f"Enter student's {fact} ({', '.join(possible_values)}): ").strip().lower()
        if user_input in possible_values:
            user_knowledge_base[fact] = user_input
            break
        else:
            print(f"Invalid input for {fact}. Please choose from {', '.join(possible_values)}.")

print("\nUser Knowledge Base:")
print(user_knowledge_base)

# Run the expert system with user-provided facts
user_recommendations = run_expert_system(user_knowledge_base, rules)

print("\n--- Recommendations Based on Your Input ---")
if user_recommendations:
    for rec in user_recommendations:
        print(f"- {rec}")
else:
    print("No specific recommendations based on the provided student profile.")

print("\nTry running this cell again with different combinations of inputs to explore various educational scenarios!")
```
